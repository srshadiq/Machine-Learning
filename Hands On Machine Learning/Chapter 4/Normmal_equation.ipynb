```python
import numpy as np
X= 2*np.random.rand(100,1)
Y= 4+3*X+np.random.rand(100,1)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, 4 + 3 * X, color='red', label='True line (no noise)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
X_b = np.c_[np.ones((100, 1)), X]
theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ Y
theta
```

```python
plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, X_b @ theta, color='red', label='Prediction Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
